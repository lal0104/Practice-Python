```python
import matplotlib.pyplot as plt

def fibonacci(n):
    if n<=1:
        ans=n
    else:
        ans=fibonacci(n-1)+fibonacci(n-2)
    return ans
no_terms=int(input("enter number of terms:"))
s=[]
for num in range(no_terms):
    print(fibonacci(num))
    s.append(fibonacci(num))
    
plt.plot(s,'bo-')
plt.xlabel("index")
plt.ylabel("fibonacci_value")
plt.title("fibonacci_sequence")
plt.grid(True)
plt.show()
```
